```python
import pandas as pd
import numpy as np
```

```python
for experiment in [
    "run_advanced_additional_datasets_small",
    "run_base_additional_datasets_small",
]:
    performed = pd.read_csv(f"../results/{experiment}/performed_runs.txt", header=None)
    scores = pd.read_json(
        f"../results/{experiment}/results.json", orient="records", lines=True
    )

    score = scores.params.apply(pd.Series)
    scores = pd.concat([scores, score], axis=1)

    scores.drop(
        columns=["params", "elapsed", "estimated_operations", "train", "test"],
        inplace=True,
    )

    labels = (
        scores["dataset_name"]
        + "__"
        + scores["oversampler_name"]
        + "__"
        + scores["model_name"]
        + "__"
        + scores["n_neighbors"].astype(str)
        + "__"
        + scores["take"].astype(str)
    )

    labels.drop_duplicates(inplace=True)
    labels.reset_index(drop=True, inplace=True)
    labels = labels.to_numpy()

    print(f"Experiment: {experiment}")
    print("Labels count:", len(labels))
    print("Performed count:", len(performed[0]))
    print("Performed labels in labels count:", performed[0].isin(labels).sum())
    
    unmatched_labels = labels[~np.isin(labels, performed[0])]
    matched_performed = performed[0][performed[0].isin(labels)]
    # matched_performed.to_csv(
    #     f"../results/{experiment}/performed_runs.txt", index=False, header=False
    # )
```

```python
for experiment in [
    "run_advanced_ignore_additional_datasets_small",
    "run_base_ignore_additional_datasets_small",
]:
    performed = pd.read_csv(f"../results/{experiment}/performed_runs.txt", header=None)
    scores_a = pd.read_json(
        f"../results/{experiment}/results_part_a.json", orient="records", lines=True
    )
    scores_b = pd.read_json(
        f"../results/{experiment}/results_part_b.json", orient="records", lines=True
    )
    scores_c = pd.read_json(
        f"../results/{experiment}/results_part_c.json", orient="records", lines=True
    )   
    
    if experiment == "run_base_ignore_additional_datasets_small":
        scores_d = pd.read_json(
            f"../results/{experiment}/results_part_d.json", orient="records", lines=True
        )
    
        scores = pd.concat([scores_a, scores_b, scores_c, scores_d], ignore_index=True)
    else:
        scores = pd.concat([scores_a, scores_b, scores_c], ignore_index=True)

    score = scores.params.apply(pd.Series)
    scores = pd.concat([scores, score], axis=1)

    scores.drop(
        columns=["params", "elapsed", "estimated_operations", "train", "test"],
        inplace=True,
    )

    labels = (
        scores["dataset_name"]
        + "__"
        + scores["oversampler_name"]
        + "__"
        + scores["model_name"]
        + "__"
        + scores["n_neighbors"].astype(str)
        + "__"
        + scores["take"].astype(str)
    )
    
    labels.drop_duplicates(inplace=True)
    labels.reset_index(drop=True, inplace=True)
    labels = labels.to_numpy()
    
    print(f"Experiment: {experiment}")
    print("Labels count:", len(labels))
    print("Performed count:", len(performed[0]))
    print("Performed labels in labels count:", performed[0].isin(labels).sum())
    
    unmatched_labels = labels[~np.isin(labels, performed[0])]
    matched_performed = performed[0][performed[0].isin(labels)]
    # matched_performed.to_csv(
    #     f"../results/{experiment}/performed_runs.txt", index=False, header=False
    # )
```
