```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

# Polish Companies

```python
with open("./results/polish_companies_results.json", "r") as file:
    data = json.load(file)
```

```python
means = {}
stds = {}
```

```python
for model, results in data.items():
    means[model] = {}
    stds[model] = {}
    for metric, values in results.items():
        means[model][metric] = np.mean(values)
        stds[model][metric] = np.std(values)
```

```python
means_df = pd.DataFrame(means).T
std_df = pd.DataFrame(stds).T
```

```python
means_df.reset_index(drop=False, inplace=True)
std_df.reset_index(drop=False, inplace=True)

means_df.rename(columns={"index": "model"}, inplace=True)
std_df.rename(columns={"index": "model"}, inplace=True)
```

```python
means_df = means_df.melt(id_vars="model", value_vars=["recalls", "precisions", "fscores"],
                        var_name="metric", value_name="value")

std_df = std_df.melt(id_vars="model", value_vars=["recalls", "precisions", "fscores"],
                        var_name="metric", value_name="value")
```

```python
sns.barplot(data=means_df, x="metric", y="value", hue="model")
plt.title("Mean values")
plt.ylabel("Score")
plt.xlabel("Label")
plt.legend(title="Model")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=std_df, x="metric", y="value", hue="model")
plt.title("Std values")
plt.ylabel("Score")
plt.xlabel("Label")
plt.legend(title="Model")
plt.tight_layout()
plt.show()
```
