```python
import os
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings

from src.results_utils import create_results_df, RESULTS_DIR
from src.plots_utils import (
    create_results_bar_plot,
    save_plot,
    create_results_point_plot,
)

TABLES_DIR = os.path.abspath(os.path.join(".", "tables"))
os.makedirs(TABLES_DIR, exist_ok=True)


def save_to_csv(df, filename):
    """Save DataFrame to CSV."""
    filepath = os.path.join(TABLES_DIR, filename)
    df.to_csv(filepath, index=False)


warnings.filterwarnings("ignore")
```

# Datasets Info

```python
dataset_info = pd.read_csv(os.path.join(RESULTS_DIR, "data_comphrasion_df.csv"))
dataset_info.head(10)
```

# Experiments Results

```python
results_df = create_results_df()

print("Number of rows in results_df:", results_df.shape[0])
```

```python
results_df = results_df.merge(
    dataset_info, left_on="dataset_name", right_on="dataset", how="left"
)
```

## Preparation

```python
results_df["param_name"] = results_df["params"].apply(
    lambda row: ", ".join(
        f"{k}={v}" for k, v in sorted(row.items(), key=lambda x: x[0])
    )
)
results_df["oversampling_hue"] = (
    results_df["oversampler_name"] + "_" + results_df["n_neighbors"].astype(str)
)

results_df.sort_values(
    ["oversampling_option", "n_neighbors", "oversampler_name"],
    inplace=True,
    ascending=False,
)
results_df.reset_index(drop=True, inplace=True)
```

```python
datasets = results_df.dataset_name.unique()
datasets
```

```python
results_df.head(10)
```

```python
agg_results_df = (
    results_df.groupby(
        [
            "dataset_name",
            "param_name",
            "oversampler_name",
            "n_neighbors",
            "oversampling_option",
        ]
    )["f1"]
    .agg(["mean", "std"])
    .reset_index()
)
```

```python
unique_methods = np.sort(agg_results_df["oversampler_name"].unique())
unique_neighbors = np.sort(agg_results_df["n_neighbors"].unique())

base_colors = sns.color_palette("husl", n_colors=len(unique_methods))
base_colors = {name: color for name, color in zip(unique_methods, base_colors)}

pallette = {}
for k in unique_neighbors:
    for name, color in base_colors.items():
        if k == 3:
            color = tuple(c * 1.25 for c in color)
        elif k == 7:
            color = tuple(c * 0.75 for c in color)
        pallette[f"{name}_{k}"] = np.clip(color, 0.0, 1.0)
```

```python
agg_results_df_ultimate = (
    results_df.groupby(
        ["dataset_name", "oversampler_name", "n_neighbors", "oversampling_option"]
    )["f1"]
    .agg(["mean", "std"])
    .reset_index()
)
```

```python
save_to_csv(agg_results_df_ultimate, "results_agg_ultimate.csv")
save_to_csv(agg_results_df, "results_agg.csv")
```

# Analysis

```python
comparison_info_detailed = agg_results_df.pivot(
    columns=["n_neighbors", "oversampling_option"],
    index=["dataset_name", "param_name", "oversampler_name"],
    values=["mean", "std"],
)
save_to_csv(comparison_info_detailed, "comparison_info_detailed.csv")
comparison_info_detailed.sort_index().head(10)
```

```python
def select_option(group):
    # If the two mean values are equal, return a row with oversampling_option set to "EQUAL"
    if group["mean"].nunique() == 1:
        return pd.Series(
            {
                "dataset_name": group["dataset_name"].iloc[0],
                "oversampler_name": group["oversampler_name"].iloc[0],
                "n_neighbors": group["n_neighbors"].iloc[0],
                "oversampling_option": "EQUAL",
            }
        )
    else:
        # Otherwise, select the row with the maximum mean value
        idx = group["mean"].idxmax()
        return group.loc[
            idx,
            ["dataset_name", "oversampler_name", "n_neighbors", "oversampling_option"],
        ]
```

```python
comparison_info_detailed_exploded = comparison_info_detailed.stack(
    level=[1, 2]
).reset_index()

max_options_detailed = (
    comparison_info_detailed_exploded.groupby(
        ["dataset_name", "oversampler_name", "n_neighbors", "param_name"]
    )
    .apply(select_option)
    .reset_index(drop=True)
)
max_options_detailed = max_options_detailed.reset_index(drop=True)

max_options_detailed
```

```python
aggregated_stats_df_detailed = pd.DataFrame(
    max_options_detailed.groupby(["oversampler_name", "n_neighbors"])[
        "oversampling_option"
    ].value_counts()
)
aggregated_stats_df_detailed
```

```python
comparison_info_ultimate = agg_results_df_ultimate.pivot(
    columns=["n_neighbors", "oversampling_option"],
    index=["dataset_name", "oversampler_name"],
    values=["mean", "std"],
)
save_to_csv(comparison_info_ultimate, "comparison_info_ultimate.csv")
comparison_info_ultimate.sort_index().head(10)
```

```python
max_options_ultimate = (
    max_options_detailed.groupby(["dataset_name", "oversampler_name", "n_neighbors"])[
        "oversampling_option"
    ]
    .agg(
        lambda x: "EQUAL"
        if x.value_counts().get("ADVANCED", 0) == x.value_counts().get("BASIC", 0)
        else x.value_counts().idxmax()
    )
    .reset_index()
)

max_options_ultimate = max_options_ultimate.reset_index(drop=True)

max_options_ultimate
```

```python
aggregated_stats_df_ultimate = pd.DataFrame(
    max_options_ultimate.groupby(["oversampler_name", "n_neighbors"])[
        "oversampling_option"
    ].value_counts()
)
aggregated_stats_df_ultimate
```

# Charts

## Basic comparison top 3

```python
for dataset in datasets:
    df = results_df[results_df["dataset_name"] == dataset]
    df_agg = agg_results_df[agg_results_df["dataset_name"] == dataset]

    best_params = (
        df_agg.sort_values("mean", ascending=False)["param_name"].head(3).to_numpy()
    )
    ratio = dataset_info[dataset_info["dataset"] == dataset]["ratio (0 : 1)"].values[0]

    figure, axs = plt.subplots(
        nrows=3, ncols=1, figsize=(20, 12), sharey=True, sharex=True
    )
    axs = axs.flatten()

    for i, (ax, param) in enumerate(zip(axs, best_params)):
        cur_df = df[df["param_name"] == param]
        create_results_bar_plot(ax=ax, df=cur_df, param=param, pallette=pallette, i=i)

    plt.suptitle(f"F1 Score Comparison for {dataset} (Ratio: {ratio})", fontsize=20)
    plt.tight_layout()
    save_plot(
        os.path.join("basic_comparison_top_3", f"f1_score_comparison_{dataset}.png")
    )
```

## Other charts

```python
aggregated_stats_df_detailed = aggregated_stats_df_detailed.reset_index()
aggregated_stats_df_detailed["oversampling_hue"] = (
    aggregated_stats_df_detailed["oversampler_name"]
    + "_"
    + aggregated_stats_df_detailed["n_neighbors"].astype(str)
)

fig, ax = plt.subplots(figsize=(12, 6))

create_results_bar_plot(
    ax=ax,
    df=aggregated_stats_df_detailed,
    param="oversampling_option",
    pallette=pallette,
    target="count",
    y_label="Count",
    bbox_right=1.22,
    round_digits=0,
)

save_plot(os.path.join("comparison", "detailed_scores.png"))
```

```python
aggregated_stats_df_ultimate = aggregated_stats_df_ultimate.reset_index()
aggregated_stats_df_ultimate["oversampling_hue"] = (
    aggregated_stats_df_ultimate["oversampler_name"]
    + "_"
    + aggregated_stats_df_ultimate["n_neighbors"].astype(str)
)

fig, ax = plt.subplots(figsize=(12, 6))

create_results_bar_plot(
    ax=ax,
    df=aggregated_stats_df_ultimate,
    param="oversampling_option",
    pallette=pallette,
    target="count",
    y_label="Count",
    bbox_right=1.22,
    round_digits=0,
)

save_plot(os.path.join("comparison", "ultimate_scores.png"))
```

```python
fig, axs = plt.subplots(
    nrows=len(unique_methods),
    ncols=len(unique_neighbors),
    figsize=(7 * len(unique_neighbors), 4 * len(unique_methods)),
    sharex=True,
    sharey=True,
)

unique_neighbors = sorted(results_df["n_neighbors"].unique())
for i, method in enumerate(unique_methods):
    for j, n_val in enumerate(unique_neighbors):
        ax = (
            axs[i, j]
            if (len(unique_methods) > 1 and len(unique_neighbors) > 1)
            else axs[max(i, j)]
        )
        subset = results_df[
            (results_df["oversampler_name"] == method)
            & (results_df["n_neighbors"] == n_val)
        ]

        # Use the pre-defined palette for the current method and n_neighbors
        palette_key = f"{method}_{n_val}"
        current_palette = {
            opt: pallette.get(palette_key)
            for opt in subset["oversampling_option"].unique()
        }
        plot_legend = i == 0 and j == len(unique_neighbors) - 1

        create_results_point_plot(
            df=subset,
            method=method,
            n_val=n_val,
            ax=ax,
            pallette=current_palette,
            plot_legend=plot_legend,
            target_y="f1",
            scale=0.8,
            capsize=0.1,
        )

plt.suptitle(
    "Mean F1 Score Comparison by Inbalance Strength and Oversampling Method",
    fontsize=20,
    y=1.00,
)
plt.tight_layout()

save_plot(os.path.join("comparison", "mean_f1_over_imbalance_strength.png"))
```
